## A-1

```python
# Answers to A-1

import pandas as pd
df1 = pd.read_csv('HW3_DataA1.csv')  # Read the file into a DataFrame

# Display the first 5 rows of the DataFrame
print("First 5 rows of df1:")
display(df1.head(5))

# Display the information of the DataFrame
print("Information of df1:")
display(df1.info())


# Print the number of columns and rows of the DataFrame
rows, colums = df1.shape
print("Number of rows:", rows)
print("Number of columns:", colums)


# Identify columns with missing data
missing_columns = df1.columns[df1.isnull().any()].tolist()
print("Missing data columns:", missing_columns)


# Fill missing data with mean or mode as appropriate
for column in missing_columns:
    if df1[column].dtype == 'object':
        df1[column].fillna(df1[column].mode()[0], inplace=True)
    else:
        df1[column].fillna(df1[column].mean(), inplace=True)
```

```python
# Answers to A-3

from sklearn.preprocessing import LabelEncoder, StandardScaler
df_encoded = df1.copy()  #to avoid modifying the original
cat_cols = df_encoded.select_dtypes(include=['object']).columns

for column in cat_cols:
    label_encoder = LabelEncoder()
    df_encoded[column] = label_encoder.fit_transform(df_encoded[column])

num_cols = df_encoded.select_dtypes(include=['int64', 'float64']).columns  # Standardize numeric columns using StandardScaler
scaler = StandardScaler()
df_encoded[num_cols] = scaler.fit_transform(df_encoded[num_cols])
display(df_encoded)
```
